```python
##Following is the distribution of 32 Discordant TP PIDs after manual decision
epsilon_pid = [
    10481049, 10741001, 10091013, 10221012, 10221013, 10421031, 10481007,
    1115100
]
Iota_pid = [10011046]
Gamma_pid = [30051002,30051018,30051020]
Alpha_PID = [10891242, 11171093, 11671007]
Delta_PID = [10891297, 92101013, 92161002]
Beta_PID = [92111002, 92161003]
Lambda = [30061026]
Other = [10251010
         ]  ### Note :'11361049','10351009' from ph3 already submitted as other
Doubt_pid = [
    92091002, 11361049, 10351009, 12061006, 12061010, 12221015, 30061016,
    30061014
]  ### Note :'11361049','10351009' from ph3 and ,'10891081 from ph2 already submitted
Doubt_pid_tobe_submitted_other = [10391003]  ##Ph2
Doubt_pid_submitted_epsilon = [10891081]  ##Ph2

#Doubt_pid_submitted_other = [11361049, 10351009]  ##Ph3
#Doubt_pid_submitted_alpha = [12061006]  ##Ph3
all_pid = epsilon_pid + Iota_pid + Gamma_pid + Alpha_PID + Delta_PID + Beta_PID + Lambda + Other + Doubt_pid_tobe_submitted_other + Doubt_pid_submitted_epsilon + Doubt_pid
len(all_pid)
```

```python
##Take manual Decison for these discordant tp PIDs
#STEP1: create a final varinat call column having one varinat val for each PID
df_all_variant_conf_aftermanual_decision = df_all_variant.copy()

## Make a df with PID and corresponding variant
df_pid_variant = df_all_variant.sort_values(['PID', 'visit']).groupby(
    'PID', as_index=False).Variant.first()

## Now add variant column with single variant value for each PID
df_all_variant_conf_aftermanual_decision_fin = pd.merge(
    df_all_variant_conf_aftermanual_decision,
    df_pid_variant,
    how='left',
    left_on='PID',
    right_on='PID')

##Copy above column to a new column named "final_var_call"
df_all_variant_conf_aftermanual_decision_fin[
    'final_var_call'] = df_all_variant_conf_aftermanual_decision_fin[
        'Variant_y']

df_variant_final = df_all_variant_conf_aftermanual_decision_fin.drop(
    'Variant_y', axis=1).rename(columns={'Variant_x': 'Variant'})

df_variant_final['final_var_call'] = np.where(
    df_variant_final['PID'].isin(Iota_pid), 'Iota',
    df_variant_final['final_var_call'])

df_variant_final['final_var_call'] = np.where(
    df_variant_final['PID'].isin(Alpha_PID), 'Alpha',
    df_variant_final['final_var_call'])

df_variant_final['final_var_call'] = np.where(
    df_variant_final['PID'].isin(Beta_PID), 'Beta',
    df_variant_final['final_var_call'])

df_variant_final['final_var_call'] = np.where(
    df_variant_final['PID'].isin(Gamma_pid), 'Gamma',
    df_variant_final['final_var_call'])

df_variant_final['final_var_call'] = np.where(
    df_variant_final['PID'].isin(Delta_PID), 'Delta',
    df_variant_final['final_var_call'])

df_variant_final['final_var_call'] = np.where(
    df_variant_final['PID'].isin(Lambda), 'Lambda',
    df_variant_final['final_var_call'])

df_variant_final['final_var_call'] = np.where(
    df_variant_final['PID'].isin(Other), 'Other',
    df_variant_final['final_var_call'])

#df_variant_final['final_var_call'] = np.where(
#    df_variant_final['PID'].isin(Doubt_pid_submitted_other), 'Other',
#    df_variant_final['final_var_call'])
#
df_variant_final['final_var_call'] = np.where(
    df_variant_final['PID'].isin(Doubt_pid_submitted_epsilon), 'Epsilon',
    df_variant_final['final_var_call'])

#df_variant_final['final_var_call'] = np.where(
#    df_variant_final['PID'].isin(Doubt_pid_submitted_alpha), 'Alpha',
#    df_variant_final['final_var_call'])
#
df_variant_final['final_var_call'] = np.where(
    df_variant_final['Variant'].isin(['R1', 'Theta', 'Eta']), 'Other',
    df_variant_final['final_var_call'])

df_variant_final['final_var_call'] = np.where(
    df_variant_final['PID'].isin(Doubt_pid_tobe_submitted_other), 'Other',
    df_variant_final['final_var_call'])

df_variant_final['final_var_call'] = np.where(
    df_variant_final['PID'].isin(Doubt_pid), '0',
    df_variant_final['final_var_call'])

print('PID',df_variant_final.PID.nunique())

print('SID',df_variant_final.SID.nunique())
```

```python
#Lets see how many PIDs have more than one tp
PID_w_ontp = []
PID_w_multitp = []
df_brii_US_upload_fin = pd.DataFrame()
for pid in df_brii_US_upload_ready.PID.unique():
    #print(pid)
    df_cons = df_brii_US_upload_ready[df_brii_US_upload_ready['PID']==pid]
    visit_day_present = df_cons['visit'].values
    visit_day_sel = np.nanmin(visit_day_present)
    #print('visit_day_present',visit_day_present)
    #print('visit_day_seleected',visit_day_sel)
    if len(visit_day_present)<2:
        PID_w_ontp.append(pid)
    else:
        PID_w_multitp.append(pid)
        
    ## Select one tp from each PID 
    df_cons_fin = df_cons[df_cons['visit']==visit_day_sel]
    if df_cons_fin[(df_cons_fin['Dosage']=='BRII_Ph2')].shape[0]>1 :
        df_cons_fin2 = df_cons_fin[df_cons_fin['primstr']=='NPH']
        
        if df_cons_fin2.shape[0]>1:
            df_cons_fin21 = df_cons_fin2.iloc[0,:]
            df_brii_US_upload_fin=df_brii_US_upload_fin.append(df_cons_fin21)
        else:
            df_brii_US_upload_fin=df_brii_US_upload_fin.append(df_cons_fin2)
            
    elif df_cons_fin[(df_cons_fin['Dosage']=='BRII_Ph3')].shape[0]>1 :
        df_cons_fin3 = df_cons_fin[df_cons_fin['primstr']=='NAS']
        df_brii_US_upload_fin=df_brii_US_upload_fin.append(df_cons_fin3)
    else:
        df_brii_US_upload_fin=df_brii_US_upload_fin.append(df_cons_fin)

print(' PIDS with only one tp',len(PID_w_ontp) )
print(' PIDS with multi tp',len(PID_w_multitp) )

#
```
